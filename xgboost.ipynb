```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
data1 = data.copy()
data1 = data1.apply(LabelEncoder().fit_transform)
data1.head()
```

```python
x = data1.drop(['income'], axis = 1)
y = data1['income']
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 355 )
```

```python
scaler = StandardScaler()
s = scaler.fit_transform(x)
```

## XGboost

```python
from xgboost import XGBClassifier
model = XGBClassifier(objective='binary:logistic')
model.fit(x_train, y_train )
```

```python
pred = model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
acc =accuracy_score(y_train,pred)
```

## Hyperparameter Tunning on XGboost

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'learning_rate': [1, 0.5, 0.2, 0.3, 0.1, 0.01, 0.001,0.02, 0.002],
    'max_depth': [4, 5, 8, 10, 14, 15, 18, 20],
    'n_estimators':[10,20,30,50,80,100,125, 140, 150, 175, 200]
}
```

```python
grid = GridSearchCV(XGBClassifier(objective='binary:logistic'), param_grid, verbose =3)
```

```python
grid.best_params_
```
